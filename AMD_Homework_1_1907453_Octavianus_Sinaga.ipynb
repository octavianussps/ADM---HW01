```python
# Problem 3.1
# Swap Case Function
def swap_case(s):
    # Variable to return as new string
    result = ""
    for i in s:
        # Change the lower string
        if i.islower():
            result += i.upper()
        # Change the upper string
        else:
            result += i.lower()
    # Return the result
    return result

# Main Program
if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
# Problem 3.5
# Count Substring Function
def count_substring(string, sub_string):
    # Initiate a counter
    counter = 0
    # Get the length of inputs
    len_str = len(string)
    len_sub_str = len(sub_string)
    # Check the substring occurance:
    # 1. Check each character from index 0 to the last (len_of_substring - 1) indexes (i.e: if the len(sub) = 4, check the char of string from 0 to -3)
    # 2. Slice the string to the length of substring
    # 3. Compare the substring with sliced string, if match add the counter
    for i in range(0, len_str-len_sub_str+1):
        if string[i:len_sub_str+i] == sub_string:
            counter += 1
    return counter

# Main Program
if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

```python
# Problem 3.9
# Enter your code here. Read input from STDIN. Print output to STDOUT
# Design Door Mat Function
def design_door_mat(n):
    # Get the size from input
    height = int(n[0])
    width = int(n[1])
    # Assign characters of symbol for design 
    central = '.|.'
    edge = '-'
    # Print the upper design (from height 0 to half of height):
    # The central in accordance to 2i+1 as the indexes increase with edges as the other character
    for i in range(0, height//2):
        print((central*(2*i+1)).center(width,edge))
    # Print the center of the door:
    # In center is word WELCOME, and another character will be edges
    print('WELCOME'.center(width,edge))
    # Print the lower design:
    # 1. Decreasing from half of height to 0 with decreament factor 1
    # 2. The central in accordance to 2i+1 as the indexes decrease with edges as the other character
    for i in range(height//2-1,-1,-1):
        print((central*(2*i+1)).center(width, edge))

# Main Program
if __name__ == '__main__':
    # Input
    size = input().split(" ")
    # Call design door mat function
    design_door_mat(size)
```

```python
# Problem 3.13
# Minion Game Function
def minion_game(string):
    # List of Substrings
    sustring_list = []
    # Initiate stuart and kevin's score
    kevin = 0
    stuart = 0
    # Check for the string
    for i in range(0, len(string)):
        # Point mark
        point = len(string) - i
        # add kevin's score for the word started in vowels
        if string[i] in ['A', 'E', 'I', 'O', 'U']:
            kevin += point
        # add stuart's score for the word started consonants
        else: 
            stuart += point
    # Print the result
    if stuart > kevin:
        result = "Stuart {:d}".format(stuart)
    elif stuart < kevin:
        result = "Kevin {:d}".format(kevin)
    else:
        result = "Draw"
    return print(result)


if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
# Problem 4.10
# Mutation Function
def set_mutation(a, num):
    # Check the string for each input
    for i in range(num):
        # Declare the command an operand from input given
        command = (input().split())[0]
        operand =  set(map(int, input().split()))
        # Apply the operation for each case
        if command == "update":
            a |= operand
        elif command == "intersection_update":
            a &= operand
        elif command == "difference_update":
            a -= operand
        elif command == "symmetric_difference_update":
            a ^= operand
    # Return the sum from mutation processes
    return  sum(a)

# Main Program
if __name__ == '__main__':
    # Input
    input()
    set_a = set(map(int, input().split()))
    n = int(input())
    # Call mutation function and show the result
    mutation = set_mutation(set_a, n)
    print(mutation)
```

```python
# Problem 5.1
# Import Library
import collections as clt

# Colection Counter Function
def col_count (num, c):
    #initiate income
    income = 0
    # Check input 
    for i in range(num):
        order = list(map(int, input().split()))
        # check the size and price
        if c[order[0]] != 0:
            # If the size available, get the price tobe calculated into income
            # Decrease the number of picked size
            income += order[1]
            c[order[0]] -= 1
    #Return the money earned
    print(income) 

# Main Program
if __name__ == '__main__':
    # Input
    input()
    count = clt.Counter(list(map(int, input().split())))
    n = int(input())
    # Call collection counter function
    col_count(n, count)
```

```python
# Problem  5.4
# Import Library
from collections import OrderedDict

#Ordered Dictionary Fuction
def ordered_dict (num, od):
    # Check the input
    for i in range(num):
        rows = input().split()
        # get item from the input
        item = ' '.join(rows[:-1])
        # get price from input
        price = int(rows[-1])
        #Calculate the price, for item listed on od
        if item in od:
            od[item] += price
        else:
            od[item] = price
    # Show the item and total price
    for j in od:
        print("{:} {:}".format(j, od[j]))

# Main Program
if __name__ == '__main__':
    # Input
    ordered_dictionary = OrderedDict()
    n = int(input())
    # Call ordered dictionary function
    ordered_dict(n, ordered_dictionary)
```

```python
# Problem 5.6
# Import Library
from collections import deque

# Deque Function
def col_deque(num, d):
    #Check the inputs
    for i in range (num):
        # assign the fuction accordingly
        rows = input().split()
        if rows[0] == "append":
            d.append(int(rows[1]))
        elif rows[0] == "appendleft":
            d.appendleft(int(rows[1]))
        elif rows[0] == "pop":
            d.pop()
        elif rows[0] == "popleft":
            d.popleft()
    # Show the result
    print(*d)

# Main Program
if __name__ == '__main__':
    # Input
    deq = deque()
    n = int(input())
    # Call deque function
    col_deque(n, deq)
```

```python
# Problem 5.7: Piling Up
# Import Library
from collections import deque

# Piling up Function
def pil_up(num):
    for i in range(num):
        input()
        # initiate deque, atrray and checker
        d = deque(list(map(int, input().split())))
        arr = []
        yes = True
        while len(d) != 0:
            a = d[0]
            poss = True
            if d[-1] > a:
                a = d[-1]
                poss = False
            if poss == True:
                d.pop()
            else:
                d.popleft()
            if arr == []:
                arr.append(a)
            else:
                if arr[-1] < a:
                    yes = False
                    break
                else:
                    arr.append(a)
        if yes == True:
            print("Yes")
        else:
            print("No")

# Main Program
if __name__ == '__main__':
    # Input
    n = int(input())
    # Call deque function
    pil_up(n)
```

```python
# Problem 5.8: Company Logo
#!/bin/python3

import math
import os
import random
import re
import sys
import collections as clt

# Company Logo Function
def logo (c):
    # Define logic to sort the counter of input string
    func = lambda x: (-1 * x[1], x[0])
    # sort them
    sorted_c = sorted(c, key = func)
    # Print the letter and occurance
    for i in range(3):
        print(sorted_c[i][0], sorted_c[i][1])

# Main Program
if __name__ == '__main__':
    # Input
    s = input()
    counter = clt.Counter(s).most_common()
    # Call the logo function
    logo(counter)
```

```python
# Problem 6.2
#!/bin/python3

import math
import os
import random
import re
import sys
from datetime import datetime as dt

# Time Delta Function
def time_delta(t1, t2):
    # Set the format
    frmt = "%a %d %b %Y %H:%M:%S %z"
    # Calculate the delta using and count the total time in second
    result = int(abs(dt.strptime(t1, frmt)-dt.strptime(t2, frmt)).total_seconds())
    return result

# Main Program
if __name__ == '__main__':
    # Input
    t = int(input().strip())
    # Input the t1 and t2
    for t_itr in range(t):
        t_1 = input().strip()
        t_2 = input().strip()
        delta= time_delta(t_1, t_2)
        print(delta)
```

```python
# Problem 11.1: XML 1
import sys
import xml.etree.ElementTree as etree

# Get Attribute Function
def get_attr_number(node):
    return len(node.attrib) + sum(get_attr_number(sub) for sub in node)

#Main Program 
if __name__ == '__main__':
    sys.stdin.readline()
    xml = sys.stdin.read()
    tree = etree.ElementTree(etree.fromstring(xml))
    root = tree.getroot()
    print(get_attr_number(root))
```

```python
# Problem 11.2 :
import xml.etree.ElementTree as etree
# Initiate global value
maxdepth = 0
# Depth Function
def depth(elem, level):
    global maxdepth
    if (level == maxdepth):
        maxdepth += 1
    
    for sub in elem:
        depth(sub, level + 1)
    
#Main Program
if __name__ == '__main__':
    n = int(input())
    xml = ""
    for i in range(n):
        xml =  xml + input() + "\n"
    tree = etree.ElementTree(etree.fromstring(xml))
    depth(tree.getroot(), -1)
    print(maxdepth)
```

```python
# Problem 2.2: Kangaroo
#!/bin/python3

import math
import os
import random
import re
import sys

# Kangaroo Function 
def kangaroo(x1, v1, x2, v2):
    # If the same speed
    if v1 == v2 and x1 == x2:
        result = "YES"
    elif v1 == v2:
        result = "NO"
    # For different speed
    t = (x2 - x1) / (v1 - v2)
    if (t == round(t) and t > 0):
        result = "YES"
    else:
        result = "NO"
    return result
    
#Main Program    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    x1V1X2V2 = input().split()

    x1 = int(x1V1X2V2[0])

    v1 = int(x1V1X2V2[1])

    x2 = int(x1V1X2V2[2])

    v2 = int(x1V1X2V2[3])

    result = kangaroo(x1, v1, x2, v2)

    fptr.write(result + '\n')

    fptr.close()
```
